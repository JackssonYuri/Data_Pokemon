```python
import pandas as pd
```

```python
dados = pd.read_csv('pokemon.csv') #lê o arquivo csv
```

```python
display(dados) #vizualizar a base de dados
```

```python
#Para vizualizar a quantidade de pokémon por tipo primário

import matplotlib.pyplot as plt

total_tipo = dados['type1'].value_counts()
total_tipo.plot(kind='bar')
plt.xlabel('Tipos Primários') #plot em x
plt.ylabel('Total de Pokémon') #plot em y
plt.title('Total de Pokémon por Tipo Primário') #titulo

# Adiciona os rótulos encima de cada coluna
for i, v in enumerate(total_tipo):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.show()

# Para vizualizar a quantidade de pokémon por tipo primário, somo cada tipo primário e então no gráfico utilizo no eixo x os rótulos de tipo para ficar mais fácil de vizualizar
# Como é possível ver há muitos pokémons do tipo água(water), mas bem poucos do tipo voador(flying, levando em conta tipo primário)
```

```python
# Verificando a quantidade de Pokémon por geração
qtd_pokemon_por_geracao = dados['generation'].value_counts()
print(qtd_pokemon_por_geracao)

```

```python
# Mostrar essa quantidade em gráfico
qtd_pokemon_por_geracao.plot(kind='bar')
plt.xlabel('Geração') #plot em x
plt.ylabel('Total de Pokémon') #plot em y
plt.title('Total de Pokémon por Geração') #titulo

# Adiciona os rótulos encima de cada coluna
for i, v in enumerate(qtd_pokemon_por_geracao):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.show()


# A partir do gráfico de barras abaixo percebemos a quantidade de pokémons a cada geração, sendo assim, na geração 5, 1 e 3, respectativamente, são os que mais tiveram pokémons
# É possível verificar que a quantidade de pokémon diminuiu conforme as gerações foram passando
```

```python
#Para vizualizar a quantidade de pokémon por tipo secundário

contagem_tipo_secundario = dados['type2'].value_counts()
contagem_tipo_secundario.plot(kind='bar')
plt.xlabel('Tipo Secundário') #plot em x
plt.ylabel('Total de Pokémon') #plot em y
plt.title('Total de Pokémon por Tipo Secundário') #titulo

# Adiciona os rótulos encima de cada coluna
for i, v in enumerate(contagem_tipo_secundario):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.show()

# Para vizualizar a quantidade de pokémon por tipo secundário, somo cada tipo secundário e então no gráfico utilizo no eixo x os rótulos de tipo para ficar mais fácil de vizualizar
# Como é possível ver há muitos pokémons do tipo voador(flying), mas bem poucos do tipo normal(levando em conta tipo secundário)
```

```python
# Verificando a média de ataque por geração
media_ataque_por_geracao = dados.groupby('generation')['attack'].mean()

print(media_ataque_por_geracao)
```

```python
# Plotando a média de ataque por geração em gráfico

media_ataque_por_geracao.plot(kind='bar')
plt.xlabel('Geração') #plot em x
plt.ylabel('Média de Ataque') #plot em y
plt.title('Média de Ataque por Geração') #titulo

# Adiciona os rótulos encima de cada coluna
for i, v in enumerate(media_ataque_por_geracao):
    plt.text(i, v, str(round(v, 2)), ha='center', va='bottom') # 2 casas decimais
plt.show()

# Vendo o gráfico da média de ataque por geração dá para notar que o ataque foi aumentando mesmo que pouco durante as gerações
```

```python
# Verificando a média de defesa por geração
media_defesa_por_geracao = dados.groupby('generation')['defense'].mean()

for geracao, media in media_defesa_por_geracao.items():
    print(f"Geração {geracao}: {media:.2f}")
```

```python
# Plotando a média de defesa por geração em gráfico

media_defesa_por_geracao.plot(kind='bar')
plt.xlabel('Geração') #plot em x
plt.ylabel('Média de Defesa') #plot em y
plt.title('Média de Defesa por Geração') #titulo

# Adiciona os rótulos encima de cada coluna
for i, v in enumerate(media_defesa_por_geracao):
    plt.text(i, v, str(round(v, 2)), ha='center', va='bottom') # 2 casas decimais
plt.show()

#A defesa também aumentou durante as gerações
```

```python
# Filtro apenas os Pokémon lendários
pokemon_lendarios = dados[dados['is_legendary'] == 1]

# Calculo a quantidade de Pokémon lendários por geração
qtd_lendarios_por_geracao = pokemon_lendarios['generation'].value_counts()

for geracao, total in qtd_lendarios_por_geracao.items():
    print(f"Geração {geracao}: {total}")
```

```python
# Vendo apenas os pokémons lendários

# Filtrar apenas os Pokémon lendários
pokemon_lendarios = dados[dados['is_legendary'] == True]

# Calcular a quantidade de Pokémon lendários por geração
qtd_lendarios_por_geracao = pokemon_lendarios['generation'].value_counts()

# Ordenar os valores por geração
qtd_lendarios_por_geracao = qtd_lendarios_por_geracao.sort_index()

# Definir o tamanho da figura
plt.figure(figsize=(10, 6))
# Criar o gráfico de barras
plt.bar(qtd_lendarios_por_geracao.index, qtd_lendarios_por_geracao, color = 'blue')
plt.xlabel('Geração')
plt.ylabel('Quantidade de Pokémon Lendários')
plt.title('Quantidade de Pokémon Lendários por Geração')

plt.bar_label(plt.bar(qtd_lendarios_por_geracao.index, qtd_lendarios_por_geracao), labels=qtd_lendarios_por_geracao, label_type='center', color='black', fontsize=8)


plt.show()

# Vizualizando melhor os pokémons lendário por geração, percebemos que na geração 7 foi a que mais apresentou lendários(17), mais que o triplo da primeira geração(5)
```
