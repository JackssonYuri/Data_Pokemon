```python
import pandas as pd
```

```python
dados = pd.read_csv('pokemon.csv') #lÃª o arquivo csv
```

```python
display(dados) #vizualizar a base de dados
```

```python
#Para vizualizar a quantidade de pokÃ©mon por tipo primÃ¡rio

import matplotlib.pyplot as plt

total_tipo = dados['type1'].value_counts()
total_tipo.plot(kind='bar')
plt.xlabel('Tipos PrimÃ¡rios') #plot em x
plt.ylabel('Total de PokÃ©mon') #plot em y
plt.title('Total de PokÃ©mon por Tipo PrimÃ¡rio') #titulo

# Adiciona os rÃ³tulos encima de cada coluna
for i, v in enumerate(total_tipo):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.show()

# Para vizualizar a quantidade de pokÃ©mon por tipo primÃ¡rio, somo cada tipo primÃ¡rio e entÃ£o no grÃ¡fico utilizo no eixo x os rÃ³tulos de tipo para ficar mais fÃ¡cil de vizualizar
# Como Ã© possÃ­vel ver hÃ¡ muitos pokÃ©mons do tipo Ã¡gua(water), mas bem poucos do tipo voador(flying, levando em conta tipo primÃ¡rio)
```

```python
# Verificando a quantidade de PokÃ©mon por geraÃ§Ã£o
qtd_pokemon_por_geracao = dados['generation'].value_counts()
print(qtd_pokemon_por_geracao)

```

```python
# Mostrar essa quantidade em grÃ¡fico
qtd_pokemon_por_geracao.plot(kind='bar')
plt.xlabel('GeraÃ§Ã£o') #plot em x
plt.ylabel('Total de PokÃ©mon') #plot em y
plt.title('Total de PokÃ©mon por GeraÃ§Ã£o') #titulo

# Adiciona os rÃ³tulos encima de cada coluna
for i, v in enumerate(qtd_pokemon_por_geracao):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.show()


# A partir do grÃ¡fico de barras abaixo percebemos a quantidade de pokÃ©mons a cada geraÃ§Ã£o, sendo assim, na geraÃ§Ã£o 5, 1 e 3, respectativamente, sÃ£o os que mais tiveram pokÃ©mons
# Ã possÃ­vel verificar que a quantidade de pokÃ©mon diminuiu conforme as geraÃ§Ãµes foram passando
```

```python
#Para vizualizar a quantidade de pokÃ©mon por tipo secundÃ¡rio

contagem_tipo_secundario = dados['type2'].value_counts()
contagem_tipo_secundario.plot(kind='bar')
plt.xlabel('Tipo SecundÃ¡rio') #plot em x
plt.ylabel('Total de PokÃ©mon') #plot em y
plt.title('Total de PokÃ©mon por Tipo SecundÃ¡rio') #titulo

# Adiciona os rÃ³tulos encima de cada coluna
for i, v in enumerate(contagem_tipo_secundario):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.show()

# Para vizualizar a quantidade de pokÃ©mon por tipo secundÃ¡rio, somo cada tipo secundÃ¡rio e entÃ£o no grÃ¡fico utilizo no eixo x os rÃ³tulos de tipo para ficar mais fÃ¡cil de vizualizar
# Como Ã© possÃ­vel ver hÃ¡ muitos pokÃ©mons do tipo voador(flying), mas bem poucos do tipo normal(levando em conta tipo secundÃ¡rio)
```

```python
# Verificando a mÃ©dia de ataque por geraÃ§Ã£o
media_ataque_por_geracao = dados.groupby('generation')['attack'].mean()

print(media_ataque_por_geracao)
```

```python
# Plotando a mÃ©dia de ataque por geraÃ§Ã£o em grÃ¡fico

media_ataque_por_geracao.plot(kind='bar')
plt.xlabel('GeraÃ§Ã£o') #plot em x
plt.ylabel('MÃ©dia de Ataque') #plot em y
plt.title('MÃ©dia de Ataque por GeraÃ§Ã£o') #titulo

# Adiciona os rÃ³tulos encima de cada coluna
for i, v in enumerate(media_ataque_por_geracao):
    plt.text(i, v, str(round(v, 2)), ha='center', va='bottom') # 2 casas decimais
plt.show()

# Vendo o grÃ¡fico da mÃ©dia de ataque por geraÃ§Ã£o dÃ¡ para notar que o ataque foi aumentando mesmo que pouco durante as geraÃ§Ãµes
```

```python
# Verificando a mÃ©dia de defesa por geraÃ§Ã£o
media_defesa_por_geracao = dados.groupby('generation')['defense'].mean()

for geracao, media in media_defesa_por_geracao.items():
    print(f"GeraÃ§Ã£o {geracao}: {media:.2f}")
```

```python
# Plotando a mÃ©dia de defesa por geraÃ§Ã£o em grÃ¡fico

media_defesa_por_geracao.plot(kind='bar')
plt.xlabel('GeraÃ§Ã£o') #plot em x
plt.ylabel('MÃ©dia de Defesa') #plot em y
plt.title('MÃ©dia de Defesa por GeraÃ§Ã£o') #titulo

# Adiciona os rÃ³tulos encima de cada coluna
for i, v in enumerate(media_defesa_por_geracao):
    plt.text(i, v, str(round(v, 2)), ha='center', va='bottom') # 2 casas decimais
plt.show()

#A defesa tambÃ©m aumentou durante as geraÃ§Ãµes
```

```python
# Vendo apenas os pokÃ©mons lendÃ¡rios

# Filtrar apenas os PokÃ©mon lendÃ¡rios
pokemon_lendarios = dados[dados['is_legendary'] == True]

# Calcular a quantidade de PokÃ©mon lendÃ¡rios por geraÃ§Ã£o
qtd_lendarios_por_geracao = pokemon_lendarios['generation'].value_counts()

# Ordenar os valores por geraÃ§Ã£o
qtd_lendarios_por_geracao = qtd_lendarios_por_geracao.sort_index()

# Definir o tamanho da figura
plt.figure(figsize=(10, 6))
# Criar o grÃ¡fico de barras
plt.bar(qtd_lendarios_por_geracao.index, qtd_lendarios_por_geracao, color = 'blue')
plt.xlabel('GeraÃ§Ã£o')
plt.ylabel('Quantidade de PokÃ©mon LendÃ¡rios')
plt.title('Quantidade de PokÃ©mon LendÃ¡rios por GeraÃ§Ã£o')

plt.bar_label(plt.bar(qtd_lendarios_por_geracao.index, qtd_lendarios_por_geracao), labels=qtd_lendarios_por_geracao, label_type='center', color='black', fontsize=8)


plt.show()

# Vizualizando melhor os pokÃ©mons lendÃ¡rio por geraÃ§Ã£o, percebemos que na geraÃ§Ã£o 7 foi a que mais apresentou lendÃ¡rios(17), mais que o triplo da primeira geraÃ§Ã£o(5)
```
